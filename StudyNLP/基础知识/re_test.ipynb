```python
with open("pep20.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        m = re.fullmatch('<tr class="field"><th class="field-name">(.*)</th><td class="field-body">(.*)</td>', line.strip())
        if m:
            print(m.groups())
```
